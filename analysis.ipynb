# All Graphs

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('data/res.csv', skipinitialspace=True)

data['command'] = data['command'].astype(str)
commands = sorted(data['command'].unique())

fig, ax = plt.subplots(figsize=(10,6))

cmdMap = {
    '1' : "Date & Time",
    '2' : "Uptime",
    '3' : "Memory Use",
    '4' : "Netstat",
    '5' : "Current Users",
    '6' : "Running Processes"
}

colorMap = {
    '1' : "blue",
    '2' : "orange",
    '3' : "green",
    '4' : "red",
    '5' : "purple",
    '6' : "brown"
}

for command in commands:
    command_data = data[data['command'] == command]
    command_data = command_data.sort_values(by="numThreads")
    ax.plot(
        command_data['numThreads'],
        command_data['averageTime'],
        label=cmdMap[command],
        marker='o',
        color=colorMap[command]
    )

ax.set_title('Average time vs. Thread Count')
ax.set_xlabel('Thread Count')
ax.set_ylabel('Average Time (seconds)')

ax.legend(title="Command")
ax.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()

plt.show()
```
